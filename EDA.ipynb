```python
# Read in datasets from countries that are predominantly English-speaking (United States, Great Britian, and Canada): 
CA_data = pd.read_csv('/content/drive/MyDrive/266 Final Project/Data/CA_youtube_trending_data.csv')
GB_data = pd.read_csv('/content/drive/MyDrive/266 Final Project/Data/GB_youtube_trending_data.csv')
US_data = pd.read_csv('/content/drive/MyDrive/266 Final Project/Data/US_youtube_trending_data.csv')
```

```python
# Concatenate all videos into a single df: 
df = pd.concat([CA_data, GB_data, US_data])
df.head()
```

```python
# Scatter Plot of 'view_count' vs. 'likes':
plt.scatter(x = df['view_count'], y = df['likes'])
# Obtain m (slope) and b(intercept) of linear regression line:
m1, b = np.polyfit(df['view_count'], df['likes'], 1)
# Add linear regression line to scatterplot:
plt.plot(df['view_count'], m1*df['view_count']+b, c = 'red')
plt.xlabel('view count')
plt.ylabel('likes')
plt.title('likes vs. view count'.title())
```

```python
# Scatter Plot of 'view_count' vs. 'comment_count':
plt.scatter(x = df['view_count'], y = df['comment_count'])
# Obtain m (slope) and b(intercept) of linear regression line:
m2, b = np.polyfit(df['view_count'], df['comment_count'], 1)
# Add linear regression line to scatterplot:
plt.plot(df['view_count'], m2*df['view_count']+b, c = 'red')
plt.xlabel('view count')
plt.ylabel('comment count')
plt.title('comment count vs. view count'.title())
```

Videos with higher numbers of views seem to be more correlated with a larger number of likes as opposed to a larger number of comments. Makes sense because it takes less work to like than to comment.

```python
# Scatter Plot of 'comment_count' vs. 'likes':
plt.scatter(x = df['comment_count'], y = df['likes'])
# Obtain m (slope) and b(intercept) of linear regression line:
m1, b = np.polyfit(df['comment_count'], df['likes'], 1)
# Add linear regression line to scatterplot:
plt.plot(df['comment_count'], m1*df['comment_count']+b, c = 'red')
plt.xlabel('comment count')
plt.ylabel('likes')
plt.title('likes vs. comment count'.title())
```

```python
# Scatter Plot of 'comment_count' vs. 'dislikes':
plt.scatter(x = df['comment_count'], y = df['dislikes'])
# Obtain m (slope) and b(intercept) of linear regression line:
m1, b = np.polyfit(df['comment_count'], df['dislikes'], 1)
# Add linear regression line to scatterplot:
plt.plot(df['comment_count'], m1*df['comment_count']+b, c = 'red')
plt.xlabel('comment count')
plt.ylabel('dislikes')
plt.title('dislikes vs. comment count'.title())
```

```python
# Convert or ensure that the publish and trending date are in datetime format:
df['trending_date']= pd.to_datetime(df['trending_date'])
df['publishedAt']= pd.to_datetime(df['publishedAt'])

# Generate a new column with the difference between the time the video was published and when it was trending
df['time_difference'] = (df['trending_date'] - df['publishedAt']).dt.total_seconds()/60/60
df.head()
```

```python
# Histogram for distribution of the time the video was published when it was trending:
plt.hist(df['time_difference'], bins=25)

# Format:
plt.xlabel('Time Until Trending (in hours)')
plt.ylabel('Frequency')
plt.title('Distribution of Time to Trending Date')
```

It appears that most videos on the trending page appeared there within 8 days of publishing.

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=12, rotation=45)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=12)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=12)
plt.title('Correlation Matrix', fontsize=14);
```

It appears that like counts are more closely linked to view count compared to dislike counts, which makes sense since an individual is more likely to watch a video repeatedly if they like it or even watch it at all if they are more likely to like it. The same relationship exists between comment count and view count.
